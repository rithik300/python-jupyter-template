### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 100)
y = x ** 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line plot
axs[0, 0].plot(x, y, color='b')
axs[0, 0].set_title('Line Chart')

# Bar plot
axs[0, 1].bar(categories, values, color=['blue', 'green', 'red', 'purple'])
axs[0, 1].set_title('Bar Chart')

# Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y, color='blue')
axs[1, 0].set_title('Scatter Plot')

# Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'orange'])
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
